```python
import ee
ee.Authenticate()
ee.Initialize()
```

```python
import geemap
```

```python
roi = ee.Geometry.Point([-0.627983, 52.074361]).buffer(1000)
# Chain multiple filters  
s2_collection = (ee.ImageCollection('COPERNICUS/S2_HARMONIZED')
    .filterDate('2022-05-01', '2023-05-01')
    .filterBounds(roi)
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20)))
```

```python
S2_minMax = s2_collection.reduce(ee.Reducer.minMax()) 
```

```python
S2_fcVis = {'bands': ["B8_min", "B4_min", "B3_min"], 'min':0, 'max': 3000}
Map = geemap.Map()
Map.center_object(roi)
Map.addLayer(S2_minMax,vis_params=S2_fcVis,name="min")
Map
```

```python
S2_minMax = s2_collection.reduce(ee.Reducer.minMax()).clip(roi) 
Map2 = geemap.Map()
Map2.center_object(roi)
Map2.addLayer(S2_minMax,vis_params=S2_fcVis,name="median")
Map2
```

```python
# Image Region Reduction
S2_minMaxReduced = S2_minMax.reduceRegion(
  reducer=ee.Reducer.mean(),
  geometry=roi,
  scale=10
)
```
